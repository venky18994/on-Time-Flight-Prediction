```
Model_Data = flightData.withColumn('CRSDepTime_Hours',(col('CRSDepTime').cast('int')/100).cast('int')).withColumn('CRSArrTime_Hours',(col('CRSArrTime').cast('int')/100).cast('int')).select('Month','DayOfWeek','CRSDepTime_Hours','CRSArrTime_Hours','UniqueCarrier','CRSElapsedTime','Origin','Dest','Distance','ArrDelay','DepDelay').withColumn('DayOfWeek',col('DayOfWeek').cast('int')).withColumn('CRSElapsedTime',col('CRSElapsedTime').cast('int')).withColumn('Distance',col('Distance').cast('int')).withColumn('label',col('ArrDelay').cast('int')).filter(col('label').isNotNull())
```

```
display(Model_Data)
```

```
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoderEstimator
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit
```

```
indexer1 = StringIndexer(inputCol='UniqueCarrier', outputCol='UniqueCarrierIndexer')
indexer2 = StringIndexer(inputCol='Origin', outputCol='OriginIndexer')
indexer3 = StringIndexer(inputCol='Dest', outputCol='DestIndexer')
#encoder = OneHotEncoderEstimator(inputCols=['UniqueCarrier','Origin','Dest','CRSDepTime_Hours','CRSArrTime_Hours'], outputCols=['UniqueCarrierIndexerVec','OriginIndexerVec','DestIndexerVec','CRSDepTime_HoursVec','CRSArrTime_HoursVec'])
encoder = OneHotEncoderEstimator(inputCols=['UniqueCarrierIndexer','OriginIndexer','DestIndexer','CRSDepTime_Hours','CRSArrTime_Hours'], outputCols=['UniqueCarrierIndexerVec','OriginIndexerVec','DestIndexerVec','CRSDepTime_HoursVec','CRSArrTime_HoursVec'])
assembler = VectorAssembler(
    inputCols=['DayOfWeek','DestIndexerVec',
 'CRSArrTime_HoursVec',
 'UniqueCarrierIndexerVec',
 'CRSDepTime_HoursVec',
 'OriginIndexerVec','CRSElapsedTime','Distance'],
    outputCol='features1')
assembler2 = VectorAssembler(
    inputCols=['DayOfWeek','CRSElapsedTime','Distance'],
    outputCol='features2')
pipeline = Pipeline(stages=[indexer1,indexer2,indexer3, encoder,assembler,assembler2])
#pipeline = Pipeline(stages=[encoder,assembler])
Model_Data_Processed = pipeline.fit(Model_Data).transform(Model_Data)
```

```
trainingData = Model_Data_Processed.filter(col('Month')< 7)
#validationData = Model_Data_Processed.filter(col('Month') == 11)
testingData = Model_Data_Processed.filter(col('Month') == 12)
```

```
lr = LinearRegression(maxIter=100)
eval = RegressionEvaluator()
paramGrid = ParamGridBuilder().addGrid(lr.regParam, [0.01, 0.1, 1.0, 10.0, 100.0]).addGrid(lr.fitIntercept, [False, True]).addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]).addGrid(lr.standardization, [True, False]).addGrid(lr.featuresCol, ['features1', 'features2']).build()
tvs = TrainValidationSplit(estimator=lr, estimatorParamMaps=paramGrid, evaluator=eval, trainRatio=7/9,seed = 123)
best_model = tvs.fit(trainingData)
```

```
print(eval.evaluate(best_model.transform(testingData)))
```

```
print(eval.evaluate(best_model.transform(trainingData),{eval.metricName: "r2"}))
```
